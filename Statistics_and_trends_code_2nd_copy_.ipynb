**Categorical Plots**

```python
def plot_gender_distribution(df, column):
    plt.figure(figsize=(8, 6))
    df[column].value_counts().plot(kind='bar', color='skyblue')
    plt.title('Distribution of Responses by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
plot_gender_distribution(df, 'Gender')
```

```python
def plot_treatment_by_occupation(df, x_col, hue_col):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=x_col, hue=hue_col)
    plt.title('Treatment Received by Occupation')
    plt.xlabel(x_col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plot_treatment_by_occupation(df, 'Occupation', 'treatment')
```

```python
def plot_line_days_indoors_coping(df, x_col, y_col, hue_col):
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x=x_col, y=y_col, hue=hue_col, marker="o", palette="Set1")
    plt.title(f"{x_col} vs. {y_col} by {hue_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()
```

```python
# Function to encode categorical variables
def encode_categorical(df):
    # Create a copy of the DataFrame to avoid SettingWithCopyWarning
    df_copy = df.copy()

    for col in df_copy.select_dtypes(include=['object']).columns:
        df_copy.loc[:, col] = df_copy[col].astype('category')  # Use .loc for assignment

    df_encoded = df_copy.copy()
    for col in df_encoded.select_dtypes(['category']):
        df_encoded[col] = df_encoded[col].cat.codes  # Encode the categories
    return df_encoded
```

```python
# Encode categorical variables
df_encoded = encode_categorical(df)
```

```python
# Function to plot the correlation heatmap
def plot_correlation_heatmap(df):
    # Calculate the correlation matrix
    correlation_matrix = df.corr()

    # Create a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
    plt.title('Correlation Heatmap', fontsize=16)
    plt.xlabel('Features', fontsize=12)
    plt.ylabel('Features', fontsize=12)
    plt.show()
```

```python
# Generate the correlation heatmap
plot_correlation_heatmap(df_encoded)
```

```python
# Violin Plot: Distribution of Mood Swings by Gender
def plot_violin(data, categorical_col, response_col):
    plt.figure(figsize=(12, 6))
    sns.violinplot(x=categorical_col, y=response_col, data=data)
    plt.title(f'Violin Plot of {response_col} by {categorical_col}')
    plt.xlabel(categorical_col)
    plt.ylabel(response_col)
    plt.xticks(rotation=45)
    plt.show()
```

```python
plot_violin(df, 'Gender', 'Mood_Swings')
```
